### Dropping Id column which is useless for the analysis. 

```python
data = data.drop('Id', axis = 1)
```

### This dataset quality has 6 values. It is not good for the SVM model because it will affect the accuracy of SVM model. Therefore, I reset three levels for these 6 values. I use 0 instead of 3 and 4 to create a new value which represent the lowest level of wine. I use 1 instead of 5 and 6 to create a new value which represent the medium level of wine, and I use 2 instead of 7 and 8 to create a new value which represent the highest level of wine.

```python
data.loc[(data['quality'] == 3) | (data['quality'] == 4), 'quality'] = 0
data.loc[(data['quality'] == 5) | (data['quality'] == 6), 'quality'] = 1
data.loc[(data['quality'] == 7) | (data['quality'] == 8), 'quality'] = 2
data.quality.nunique()
```

### Next, I set up the 3 levels of quality to two groups. I select y values equal to 0 and 1 as the first group, and this group will predict the wine belongs to the lowest level or medium level. Then, I select y values equal to 0 and 2 as the second group, and this group will predict the wine belongs to the lowest level or highest level. 

```python
data1 = data.loc[(data['quality'] == 0) | (data['quality'] == 1)]
data1
```

```python
data2 = data.loc[(data['quality'] == 0) | (data['quality'] == 2)]
data2
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

### I use the first dataset data1 to run the six models logistic regression, naive bayes, decision trees, random forests, support vector machines, and K-Nearest Neighbors.

### Separating the features from the labels

```python
X1 = data1.drop('quality', axis = 1).values
```

```python
y1 = data1['quality'].values
```

### Splitting the Dataset data1 into Training and Test Sets

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.30, random_state = 10)
```

### Feature Scaling - One method to rescale data is to standardize it.

To standardize a dataset means to scale all of the values in the dataset such that the mean value is 0 and the standard deviation is 1.

```python
sc = StandardScaler()
X1_train = sc.fit_transform(X1_train)
X1_test = sc.transform(X1_test)
```

### Fitting the model to the training dataset.

Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X1_train, y1_train.ravel())
```

Decision Trees

```python
dt = DecisionTreeClassifier()
dt.fit(X1_train, y1_train.ravel())
```

Random Forests

```python
rf = RandomForestClassifier()
rf.fit(X1_train, y1_train.ravel())
```

Naive Bayes

```python
gnb = GaussianNB()
gnb.fit(X1_train, y1_train.ravel())
```

K-Nearest Neighbors

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X1_train, y1_train.ravel())
```

Support Vector Machines

```python
linear = svm.SVC(kernel='linear', C=1, decision_function_shape='ovo').fit(X1_train, y1_train)
rbf = svm.SVC(kernel='rbf', gamma=1, C=1, decision_function_shape='ovo').fit(X1_train, y1_train)
poly = svm.SVC(kernel='poly', degree=3, C=1, decision_function_shape='ovo').fit(X1_train, y1_train)
sig = svm.SVC(kernel='sigmoid', C=1, decision_function_shape='ovo').fit(X1_train, y1_train)
```

### Performing 5-Fold Cross Validation.

```python
k = 5
kf = KFold(n_splits = k)
```

Logistic Regression

```python
result = cross_val_score(lr, X1_train, y1_train, cv = kf, scoring='accuracy')
print(f' Avg accuracy:{result.mean()}')
```

Decision Tree

```python
result = cross_val_score(dt, X1_train, y1_train, cv = kf, scoring='accuracy')
print(f' Avg accuracy:{result.mean()}')
```

 Random Forests

```python
result = cross_val_score(rf, X1_train, y1_train, cv = kf, scoring='accuracy')
print(f' Avg accuracy:{result.mean()}')
```

Naive Bayes

```python
result = cross_val_score(gnb, X1_train, y1_train, cv = kf, scoring='accuracy')
print(f' Avg accuracy:{result.mean()}')
```

K-Nearest Neighbors

```python
result = cross_val_score(knn, X1_train, y1_train, cv = kf, scoring='accuracy')
print(f' Avg accuracy:{result.mean()}')
```

Support Vector Machines. SVM has 4 models. After I run them, I found svm-linear and svm-rbf have the highest accuracy. Therefore, I choose accuracy 0.961 from svm-linear as the accuracy of SVM.

```python
result = cross_val_score(linear, X1_train, y1_train.ravel(), cv = kf, scoring='accuracy')
print(f' Avg accuracy:{result.mean()}')
```

```python
result = cross_val_score(rbf, X1_train, y1_train.ravel(), cv = kf, scoring='accuracy')
print(f' Avg accuracy:{result.mean()}')
```

```python
result = cross_val_score(poly, X1_train, y1_train.ravel(), cv = kf, scoring='accuracy')
print(f' Avg accuracy:{result.mean()}')
```

```python
result = cross_val_score(sig, X1_train, y1_train.ravel(), cv = kf, scoring='accuracy')
print(f' Avg accuracy:{result.mean()}')
```
