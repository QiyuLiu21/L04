```
import pandas as pd 
# This statement imports pandas library as pd.
# pandas is a powerful data analysis and manipulation library for Python.

import numpy as np 
# This statement imports numpy as np.
# Numpy provides Provides: 
# 1. An array object of arbitrary homogeneous items 
# 2. Fast mathematical operations over arrays
# 3. Linear Algebra, Fourier Transforms, Random Number Generation

import statsmodels.formula.api as amf 
# This statement imports statsmodels.formula.api as amf.
# statsmodels.formula.api creates a Model from a formula and dataframe.

from sklearn.linear_model import LinearRegression 
# This statement imports LinearRegression from sklearn.linear_model.
# LinearRegression fits a linear model with coefficients w = (w1, …, wp) to minimize the residual 
# sum of squares between the observed targets in the dataset, and the targets predicted by the linear approximation.

from matplotlib import pyplot as plt
# This statement imports pyplot as plt.
# matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. Each pyplot 
# function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, 
# plots some lines in a plotting area, decorates the plot with labels, etc.

import seaborn as sns
# This statement imports seaborn as sns.
# Seaborn is a Python data visualization library based on matplotlib. 
# It provides a high-level interface for drawing attractive and informative statistical graphics.

sns.set()
# Set the back color of the graphy.
```

```
np.random.seed(10) # Generate random data.
# This method is called when RandomState is initialized. It can be called again to re-seed the generator. 

X = 2.5 * np.random.randn(1000) + 1.5 # Array of 100 values with mean =1.5, stddev = 2.5.
# np.random.randn(1000) returns a sample from the “standard normal” distribution. 
# This number multiply 2.5(stddev) and plus 1.5(mean) will equal to X.

res = 0.5 * np.random.randn(1000) # Generate 100 residual terms.
# Residual will be 0.5 multiply np.random.randn(1000).

y = 2 +0.3 * X +res # Actual values of Y.
# 0.3 times X plus 2 and residul will get y.
```

```
# Creat pandas dataframe to store our X and y values
df = pd.DataFrame({'X': X,'y':y})
# Two-dimensional, size-mutable, potentially heterogeneous tabular data.
# Data structure also contains labeled axes (rows and columns). 
```

```
# Calculate the mean of X and y.
xmean = np.mean(X) 
# np.mean(x) computes the arithmetic mean along the x axis.

ymean = np.mean(y)
# np.mean(y) computes the arithmetic mean along the y axis.
```

```
# Calculate the terms needed for the numator and denominator of beta.
df['xycov']=(df['X']-xmean)*(df['y']-ymean)
# The covariance between X and Y.

df['xvar']=(df['X']-xmean)**2
# The variance for a specific x. It is the square of the value of actual x minus the mean of x.
```

```
# Calculate beta and alpha.
beta =df['xycov'].sum()/df['xvar'].sum()
alpha=ymean - (beta*xmean)
print(f'alpha={alpha}')
print(f'beta={beta}')
```

```
ypred = alpha + beta* X
# We can get the regression line using alpha and beta. Given any x value, we can predict y value.
```

```
plt.figure(figsize=(12,6))
# plt.figure uses to figure instance.

plt.plot(X,ypred)
# X will be on the x axis and ypred will on the y axis. Plot the scattor plot.

plt.plot(X,y,'ro')
# Draw a line use x and y values.

plt.title('Actual vs Predicted')
# Put the Actual vs Predicted as the title of the graphy.

plt.xlabel('X')
# X axis label will show X.

plt.ylabel('y')
# Y axis label will show y.
```
